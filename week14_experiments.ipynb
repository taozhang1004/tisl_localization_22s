```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-1, num_class)
y = np.array(lst_1)

barWidth = 0.45
plt.figure(figsize=(10, 6))
plt.bar(x, y, color ='r', width = barWidth,
        edgecolor ='grey', label ='top1')
# plt.plot(x, y, linestyle="--")

for index, value in enumerate(y):
    plt.text(index, value,
             str(value))
    
plt.xlabel("obj index")
plt.xticks(x)
plt.ylabel("Top-1 acc")
plt.title('obj vs. Top-1 acc with different obj ignored')
plt.legend()
plt.grid(True)
plt.show()
```
